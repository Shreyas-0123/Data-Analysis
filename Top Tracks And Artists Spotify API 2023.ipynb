```python
#importing required Modules
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd

# Set up Spotify API credentials
SPOTIPY_CLIENT_ID = 'b2482efc6ec54f3993b123a75e7a7a75'
SPOTIPY_CLIENT_SECRET = '341d0a8c402e442ea5d73ab9ed913475'
SPOTIPY_REDIRECT_URI = 'http://localhost:3000'

# Set up Spotify API authentication
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=SPOTIPY_CLIENT_ID,
                                               client_secret=SPOTIPY_CLIENT_SECRET,
                                               redirect_uri=SPOTIPY_REDIRECT_URI,
                                               scope='playlist-read-private'))
```

## Top Hindi Tracks And Artists 2023

```python
# Get tracks from the Top Hindi Tracks of 2023 playlist
tracks = sp.playlist_tracks('37i9dQZF1DXdf5QlS7TLuX', limit=100)

# Extract relevant information from the tracks
track_data = []
for track in tracks['items']:
    track_data.append({
        'Track Name': track['track']['name'],
        'Artists': ', '.join([artist['name'] for artist in track['track']['artists']]),
        'Popularity': track['track']['popularity']
    })

# Convert data to DataFrame
df = pd.DataFrame(track_data)
```

```python
# Get artist data from the Top Hindi artists of 2023 playlist
arts = sp.playlist_tracks('37i9dQZF1DX5zs2btBfPWm', limit=100)

# Extract relevant information from the tracks
art_data = []
for art in arts['items']:
    art_data.append({
        'Track Name': art['track']['name'],
        'Artists': ', '.join([artist['name'] for artist in art['track']['artists']]),
        'Popularity': art['track']['popularity']
    })

# Convert data to DataFrame
df1 = pd.DataFrame(art_data)
```

## Top 10 Hindi Tracks  2023

```python
ht = df.sort_values(by = "Popularity",ascending = False).head(10)
ht.style.hide()
```

## Top 10 Hindi Artists

```python
ha = df1.sort_values(by = "Popularity",ascending = False).head(10)
ha1 = ha[["Artists","Popularity"]]
artists_split = ha1['Artists'].str.split(', ', expand=True)
ha1['Artist 1'] = artists_split[0]
ha1['Artist 2'] = artists_split[1]
ha1['Artist 3'] = artists_split[2]
ha1['Artist 4'] = artists_split[3]
ha1['Artist 5'] = artists_split[4]
f_ha = ha1[["Artist 1","Artist 2","Artist 3","Artist 4","Artist 5","Popularity"]]
f_ha.style.hide()
```
