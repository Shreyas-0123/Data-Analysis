## Data Analysis

## Top 5 Popular Artist

```python
top_5_p = data.sort_values(by = "artist_popularity",ascending = False).head(5)
top_5_p
```

```python
sns.barplot(x = "name",y= "artist_popularity",data = top_5_p)
plt.title("Top 5 Popular Artist",size = 15,weight = 'bold')
plt.xlabel("Artist",weight = 'bold')
plt.ylabel("Popularity",weight = 'bold')
```

## Top 5 Most Followed Artist

```python
top_5_ma = data.sort_values(by="followers",ascending = False).head(5)
top_5_ma
```

```python
sns.barplot(x = "name",y = "followers",data = top_5_ma)
plt.title("Top 5 Most Followed Artist",size = 15,weight = "bold")
plt.xlabel("Artist",weight = 'bold')
plt.ylabel("Followers",weight = 'bold')
```
